```python
import kagglehub
import pandas as pd
from kagglehub import KaggleDatasetAdapter

path = kagglehub.dataset_download("adilshamim8/student-depression-dataset")
file_name='student_depression_dataset.csv'
df = pd.read_csv(f"{path}/{file_name}")
df.drop(columns=['id'], inplace=True)
df.head()
```

```python
#group equivalent degrees into categories
degree_groups = {
    # School
    'Class 12': 'School',
    # Undergraduate
    'BSc': 'UG', 'B.Com': 'UG', 'BCA': 'UG', 'BBA': 'UG',
    'B.Tech': 'UG', 'BE': 'UG', 'B.Arch': 'UG', 'BHM': 'UG',
    'B.Pharm': 'UG', 'BA': 'UG',
    # Postgraduate
    'M.Sc': 'PG', 'M.Com': 'PG', 'MBA': 'PG', 'MA': 'PG',
    'MCA': 'PG', 'M.Tech': 'PG', 'M.Ed': 'PG', 'M.Pharm': 'PG',
    'LLM': 'PG',
    # Professional or doctoral
    'PhD': 'Doctoral', 'MD': 'Doctoral',
    'MBBS': 'Professional', 'LLB': 'Professional', 'B.Ed': 'Professional',
    'ME': 'Professional', 'MHM': 'Professional',
}
df_cut['Degree_group'] = df_cut['Degree'].replace(degree_groups)
df_cut.head()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
```

```python
dietary_habits_order = ['Unhealthy', 'Moderate', 'Healthy']
sleep_duration_order = ['Less than 5 hours','5-6 hours','7-8 hours','More than 8 hours']
```

```python
df_cut.tail()
```

```python
def encode(feature, frame, cat_order, clean): #function to make string features ordinal
  if clean:
    frame[feature] = frame[feature].astype(str).str.strip().str.replace("'", "").str.replace('"', '')
  frame[feature] = pd.Categorical(
      frame[feature],
      categories=cat_order,
      ordered=True
  )
  print(frame[feature].value_counts())
```

```python
# build a logistic regression classifier
from sklearn.preprocessing import StandardScaler

Y = df_cut['Depression']
X = df_cut.drop(columns=['Depression'])
X_train_reg, X_test_reg, Y_train_reg, Y_test_reg = train_test_split(X, Y, test_size=0.2, random_state=42)

encode('Dietary Habits', X_train_reg, dietary_habits_order, False)
encode('Sleep Duration', X_train_reg, sleep_duration_order, True)

print(X_train_reg['Dietary Habits'].cat.codes)
print(X_train_reg['Sleep Duration'].cat.codes)

X_train_reg['Dietary Habits']=X_train_reg['Dietary Habits'].cat.codes
X_train_reg['Sleep Duration']=X_train_reg['Sleep Duration'].cat.codes

encode('Dietary Habits', X_test_reg, dietary_habits_order, False)
encode('Sleep Duration', X_test_reg, sleep_duration_order, True)

X_test_reg['Dietary Habits'] = X_test_reg['Dietary Habits'].cat.codes
X_test_reg['Sleep Duration'] = X_test_reg['Sleep Duration'].cat.codes

#scale
scaler = StandardScaler()
X_train_reg_scaled = scaler.fit_transform(X_train_reg)
X_test_reg_scaled = scaler.transform(X_test_reg)

logR = LogisticRegression()
logR.fit(X_train_reg_scaled, Y_train_reg)
```

```python
logR.score(X_test_reg_scaled, Y_test_reg)
```

```python
preds = logR.predict(X_test_reg_scaled)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, balanced_accuracy_score, recall_score
```

```python
cm = confusion_matrix(Y_test_reg, preds)
print(cm)
print('accuracy:',accuracy_score(Y_test_reg, preds))
print('recall:',recall_score(Y_test_reg, preds))
print(classification_report(Y_test_reg, preds))
```
